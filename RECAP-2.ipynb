```python
fig,ax = plt.subplots(figsize=(12,7))

plt.pie(x=df["origin"].value_counts().values, labels = df["origin"].value_counts().index, 
                autopct="%.1f%%", explode=(0.1,0.2,0.3))

plt.title("A General View",color="r")
plt.show()
```

```python
df.info()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,7))
fig.suptitle("Two graphs on the same line")


ax[0].pie(x=df["origin"].value_counts().values, labels = df["origin"].value_counts().index, 
                autopct="%.1f%%", explode=(0.1,0.2,0.3))


ax[1].scatter(x=df["mpg"],y=df["horsepower"])

plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,9))
fig.suptitle("PERFECT",color="purple")

ax[0].bar(x=df.origin.value_counts().index,height=df.origin.value_counts().values)

ax[1].scatter(x=df["weight"],y=df["horsepower"])

plt.show()
```
